```python
import pandas as pd 
import numpy as np
```

```python
df_cur = pd.read_excel("witn_new_rat_ag_num_cat_catnum.xlsx")
df_cur = df_cur.loc[~df_cur["_ro_type"].str.startswith("T", na=False)]
df_cur.head(5)
```

```python
ident = df_cur["identifier"].unique()

series_length = {}
for inn in ident:
    if inn not in series_length:
        series_length[inn] = {}
    for agency in df_cur["agency"].unique():
        if agency not in series_length[inn]:
            series_length[inn][agency] = []
        temp_df = df_cur[(df_cur["identifier"] == inn) & (df_cur["agency"] == agency)]
        curr_len = len(temp_df["new_rating"])
        series_length[inn][agency] = curr_len   # there should apply the length of series of each company in cra
        
df_ser = pd.DataFrame().from_dict(series_length).T.reset_index()
df_ser.rename(columns={"index": "identifier"}, inplace=True)
df_ser[df_ser["Expert RA"] >= 10]
```

```python
from collections import defaultdict

expert_dict = defaultdict(list)

for inn in unique:
    temp_1 = df_cur[(df_cur["agency"] == "Expert RA") & (df_cur["identifier"] == int(inn))]
    temp_2 = df_last[(df_last["agency"] == "Expert RA") & (df_last["identifier"] == str(inn))]
    
    ratings_1 = list(temp_1["new_rating"].values)
    ratings_2 = list(temp_2["rating"].values)
    date_1 = [str(date)[:10] if isinstance(date, np.datetime64) else date for date in temp_1["_date"].values]
    date_2 = [str(date)[:10] if isinstance(date, np.datetime64) else date for date in temp_2["_date"].values]
    
    if inn == 7725039953 or inn == '7725039953':
        print(df_cur[(df_cur["identifier"] == 7725039953) & (df_cur["agency"] == "Expert RA")])
    
    type_ = list(temp_1["_ro_type"].values)[0] if len(temp_1) > 0 else ''
    _ro_type = list(temp_1["_ro_type"].values) + [type_] * len(temp_2)
    
    total_count = len(temp_1) + len(temp_2)

    expert_dict["identifier"].extend([inn] * total_count)
    expert_dict["new_rating"].extend(ratings_1 + ratings_2)
    expert_dict["_date"].extend(date_1 + date_2)
    expert_dict["_ro_type"].extend(_ro_type)
    expert_dict["agency"].extend(["Expert RA"] * total_count)
    expert_dict["scale"].extend(["Кредит. шкала"] * total_count)

# Создаем DataFrame
df_expert = pd.DataFrame(expert_dict)
```

```python
df_expert = df_expert.drop_duplicates()
```

```python
df_expert["identifier"] = df_expert["identifier"].astype(str)
df_cur["identifier"] = df_cur["identifier"].astype(str)
df_expert["_date"] = pd.to_datetime(df_expert["_date"])
df_cur["_date"] = pd.to_datetime(df_cur["_date"])

df_final = pd.merge(df_expert, df_cur, on=["identifier", "_date", "agency", "new_rating", "scale", "_ro_type"], how="outer")
```
