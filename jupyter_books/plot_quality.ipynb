```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Пример данных для каждого года CTMC
# years = ['1 год', '2 года', '3 года', '4 года']
# accuracy = [0.9373, 0.9094, 0.8576, 0.8201]  # Примерные значения точности
# recall = [0.80, 0.82, 0.83, 0.84]    # Примерные значения полноты
# precision = [0.78, 0.79, 0.81, 0.82] # Примерные значения точности

#DIRICHLET
years = ['1 год', '2 года', '3 года', '4 года']
Dirichlet = [0.8406, 0.7353,  0.6502, 0.5795]  # Примерные значения точности
CTMC = [0.9015,  0.7643, 0.6977, 0.6296]    # Примерные значения полноты
sec_order = [0.8520, 0.7009, 0.6386, 0.5304] # Примерные значения точности
basic_ = [ 0.6706,  0.630, 0.55144,  0.58165]
```

```python

x = np.arange(len(years))
width = 0.15

# Пастельные цвета
colors = {
    'accuracy': '#a1c9f4',    # пастельный голубой
    'recall':   '#ffb482',    # пастельный оранжевый
    'precision':'#8de5a1'     # пастельный зелёный
}

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x - width, Dirichlet, width, label='MC1/Dirichlet', color=colors['accuracy'])
ax.bar(x, CTMC, width, label='MC1/CTMC', color=colors['recall'])
ax.bar(x + width,sec_order, width, label='MC 2 order', color=colors['precision'])
ax.bar(x + 2*width,basic_, width, label='basic_', color='pink')
# Настройки осей и легенды
ax.set_xlabel('Горизонт прогноза')
ax.set_ylabel('Значение метрик')
ax.set_title('Accuracy по годам')
ax.set_xticks(x)
ax.set_xticklabels(years)
ax.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
