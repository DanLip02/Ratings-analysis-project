```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Данные из таблицы
data = np.array([
    [0.9054, 0.0838, 0.0108],  # AAA-AA
    [0.079,  0.8421, 0.0789],  # A-BBB
    [0.0118, 0.0476, 0.9406]   # D
])


# Названия строк и столбцов
labels = ['A', 'B', 'D']

# Создаем фигуру
plt.figure(figsize=(8, 6))

# Строим тепловую карту
ax = sns.heatmap(
    data,
    annot=True,
    linewidths=0.5,        # Границы между ячейками
        xticklabels=labels,
        yticklabels=labels,
    )

# Настройка заголовка и меток
plt.title("Тепловая карта переходных вероятностей", fontsize=14)
plt.xlabel("Конечное состояние", fontsize=12)
plt.ylabel("Начальное состояние", fontsize=12)

# Корректировка макета
# plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Данные из таблицы
data = [
    [0.9054, 0.0838, 0.0108],  # AAA-AA
    [0.079,  0.8421, 0.0789],  # A-BBB
    [0.0118, 0.0476, 0.9406]   # D
]

# Названия строк и столбцов
labels = ['A', 'B', 'D']

# Создаем фигуру и оси
plt.figure(figsize=(8, 6))
ax = plt.gca()

# Строим тепловую карту с цветовой шкалой
heatmap = sns.heatmap(
    data,
    annot=True,
    fmt=".4f",             # Форматирование чисел
    cmap="coolwarm",       # Цветовая схема
    linewidths=0.5,
    linecolor='black',     # Цвет линий разделения
    square=True,           # Квадратные ячейки
    cbar=True,             # Цветовая шкала
    cbar_kws={'label': 'Вероятность'},  # Подпись цветовой шкалы
    xticklabels=labels,
    yticklabels=labels,
    ax=ax
)

# Настройка заголовка и меток
plt.title("Тепловая карта переходных вероятностей", fontsize=14, pad=20)
plt.xlabel("Конечное состояние", fontsize=12)
plt.ylabel("Начальное состояние", fontsize=12)

# Настройка меток на осях
ax.set_xticklabels(labels, rotation=0, fontsize=10)
ax.set_yticklabels(labels, rotation=0, fontsize=10)

# Автоматическая регулировка отступов
plt.tight_layout()

# Показать график
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Преобразуем в массив
data = np.array([
    [0.9054, 0.0838, 0.0108],
    [0.079,  0.8421, 0.0789],
    [0.0118, 0.0476, 0.9406]
])

labels = ['A', 'B', 'D']

fig = plt.figure(figsize=(8, 6))
sns.heatmap(data, annot=True, fmt='.3f', linewidths=.5, annot_kws={"size":11})
plt.title("Тепловая карта переходных вероятностей", fontsize=14)
plt.xlabel("Конечное состояние", fontsize=12)
plt.ylabel("Начальное состояние", fontsize=12)
fig.show()
```
